##**데이터와 데이터 분석**


###**Data 란?**

####**실무자의 관점에서 데이터란?**
사전적인 정의 : 데이터는 관측이나 사건에 대한 정량적인 수치적인 정보를 데이터라고 한다.

###**데이터 분석의 발달 과정**

데이터는 한 가지 결과에 대한 값을 다루는 것이 아니라 여러가지 현상, 트렌드를 종합적으로 모아서 수치화 시키는 것을 기본적으로 데이터라고 한다.
데이터 분석을 한다는 것은 어떤 현상을 수학적으로 접근해서 풀어서 누군가에게 보여주겠다라는 뜻이다.

데이터 분석 초반에는 통계학적으로 접근하였다. -> 전통 통계
분석하려는 모집단의 일부를 추출한 표본(샘플)의 통계량을 조사하는 방법으로 분석을 하였다.
샘플의 평균과 합 등을 가지고 모집단을 추정하는 방법으로 진행하였다.
따라서 데이터 분석을 하려면 전통 통계에 대하여 어느 정도 알고 있어야 한다.

2000년대 IT 기술의 발달로, 모집단의 데이터 수집하고 저장하고 처리 할수 있게 되었다.
따라서 전통 통계에 데이터 마이닝이라는 관점이 실무에 적용되기 시작하였다.
이는 데이터의 일부를 추출하여 모집단을 추정하였는데 지금은 전체 데이터를 한꺼번에 쓸 수 있는 기술이 생겨서 전체 데이터 일부를 가져와서 컴퓨터에게 학습을 시켜서, 데이터로부터 규칙과 수식을 찾아내어 데이터를 검증하는 데이터 마이닝을 활용 => 기계학습

따라서 이제는 기계학습을 떼놓을 수 없게 되었다.

분석가들이 python, R과 같은 언어를 사용하게 되었다.
R은 전통 통계를 하기 위한 언어를 발전 시킨 언어이다.
표 이미지 뿐만 아니라 비정형 데이터인 그림과 소리를 다루기 시작하면서 R로 다루기에는 한계가 있기 때문에 파이썬을 이용하기 시작하였다.

현재는 공공데이터를 웹에서 수집하여 회사의 데이터와 붙여서 사용하거나, 직접 데이터를 수집하여 분석하여 인사이트를 얻게 되어 확장성이 좋은 파이썬을 사용하게 되었다.

데이터분석에서 시각화, 기계학습, 통계에 초점을 맞춰 분석을 하겠다 하면 R을 사용해도 된다.
데이터 수집, 처리, 이미지와 소리를 다루고 싶다면 파이썬을 선택해 사용하면 되겠다.
